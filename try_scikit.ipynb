```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
df.describe()
```

```python
df['target'] = data.target_names[data.target]
```

```python
df[df.target=='virginica'].plot.scatter(x='sepal length (cm)', y='sepal width (cm)')
```

```python
fig, ax = plt.subplots()
for name, group in df.groupby('target'):
    ax.plot(group['sepal length (cm)'], group['sepal width (cm)'], marker='o', linestyle='', ms=8, label=name)
ax.legend()
plt.show()
```

```python
import mpl_toolkits.mplot3d
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data[:, :2]
Y = iris.target
```

```python
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
```

```python
plt.figure(2, figsize=(8, 6))
plt.clf()
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Set1, edgecolors="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
```

```python
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=Y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40
)
```

```python
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.set_zlabel("3rd eigenvector")
ax.zaxis.set_ticklabels([])
```
